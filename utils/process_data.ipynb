### Combine file togather

```python


# Define paths
train_dir = '/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/ODIR-5K/ODIR-5K/Training Images'
test_dir = '/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/ODIR-5K/ODIR-5K/Testing Images'
output_dir = '/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/combine_all_image'

# Create output folder if not exists
os.makedirs(output_dir, exist_ok=True)

# Combine function
def combine_images(src_dirs, dest_dir):
    existing_files = set()

    for src in src_dirs:
        for filename in os.listdir(src):
            src_file = os.path.join(src, filename)

            # Skip non-files (e.g., system files or folders)
            if not os.path.isfile(src_file):
                continue

            dest_file = os.path.join(dest_dir, filename)

            if filename in existing_files or os.path.exists(dest_file):
                print(f" Duplicate detected, skipping: {filename}")
                continue

            shutil.copy2(src_file, dest_file)
            existing_files.add(filename)

    print(f" Successfully combined images into: {dest_dir}")
    print(f" Total images copied: {len(existing_files)}")

# Run the combination
combine_images([train_dir, test_dir], output_dir)
```

### Delete '._' this file type images

```python


# Folder path to clean
target_folder = '/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/combine_all_image'

# Scan and remove macOS '._' files
def clean_appledouble_files(folder_path):
    removed_count = 0

    for filename in os.listdir(folder_path):
        if filename.startswith("._"):
            file_path = os.path.join(folder_path, filename)
            try:
                os.remove(file_path)
                removed_count += 1
                print(f" Removed: {filename}")
            except Exception as e:
                print(f" Could not remove {filename}: {e}")

    print(f"\n Cleanup complete. Total files removed: {removed_count}")

# Run the cleaner
clean_appledouble_files(target_folder)
```

### Mapping the data. Split the disease class

```python


# Load dataset
df = pd.read_csv('/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/full_df.csv')

# Destination base folder
base_dir = '/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/preprocess_data/class_image_data'

# Disease flag → folder/class name
disease_map = {
    'N': 'Normal',
    'D': 'Diabetes',
    'G': 'Glaucoma',
    'C': 'Cataract',
    'A': 'AMD',
    'H': 'Hypertension',
    'M': 'Myopia',
    'O': 'Other'
}

# Create folder structure
os.makedirs(base_dir, exist_ok=True)
for label in disease_map.values():
    os.makedirs(os.path.join(base_dir, label), exist_ok=True)

# Map from keyword (lowercase) to flag
keyword_to_flag = {
    'normal': 'N',
    'diabetes': 'D',
    'glaucoma': 'G',
    'cataract': 'C',
    'age related macular degeneration': 'A',
    'amd': 'A',
    'hypertension': 'H',
    'myopia': 'M'
}

# Priority order of label assignment
priority_order = ['D', 'G', 'C', 'A', 'H', 'M', 'N']  # Place N lower to avoid over-labeling normal

# Assign label from diagnostic keyword string
def assign_label(keywords_str):
    keywords_str = str(keywords_str).lower()  # clean
    for flag in priority_order:
        for keyword, k_flag in keyword_to_flag.items():
            if k_flag == flag and keyword in keywords_str:
                return disease_map[flag]
    return disease_map['O']

# Copy logic
src_image_dir = '/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/combine_all_image'
class_counts = defaultdict(int)
total_images = 0

for idx, row in df.iterrows():
    for side in ['Left', 'Right']:
        img_col = f"{side}-Fundus"
        keyword_col = f"{side}-Diagnostic Keywords"
        img_name = row[img_col]

        if pd.isna(img_name):
            continue

        keywords = row[keyword_col]
        label = assign_label(keywords)

        src_path = os.path.join(src_image_dir, img_name)
        dest_path = os.path.join(base_dir, label, img_name)

        if os.path.exists(src_path):
            try:
                shutil.copy2(src_path, dest_path)
                class_counts[label] += 1
                total_images += 1
            except Exception as e:
                print(f" Failed to copy {img_name}: {e}")
        else:
            print(f" Missing file: {src_path}")

# Print summary
print("\n Dataset Preparation Complete!")
print(f"  Total Images Copied: {total_images}")
print(" Class-wise Image Count:")
for cls, count in class_counts.items():
    print(f"   - {cls}: {count}")
```

```python
import os
import shutil
import pandas as pd
import numpy as np
from pathlib import Path
import logging
from typing import Dict, List, Tuple, Optional
from collections import defaultdict
import json
from datetime import datetime

class EyeDiseaseMapper:
    """
    A comprehensive mapper class for organizing eye disease dataset images 
    into structured folders based on disease classifications.
    """
    
    def __init__(self, csv_path: str, source_image_dir: str, output_base_dir: str):
        """
        Initialize the Eye Disease Mapper
        
        Args:
            csv_path: Path to the CSV file containing dataset information
            source_image_dir: Directory containing all source images
            output_base_dir: Base directory where organized folders will be created
        """
        self.csv_path = csv_path
        self.source_image_dir = Path(source_image_dir)
        self.output_base_dir = Path(output_base_dir)
        
        # Disease mapping
        self.disease_mapping = {
            'N': 'Normal',
            'D': 'Diabetes', 
            'G': 'Glaucoma',
            'C': 'Cataract',
            'A': 'Age_related_Macular_Degeneration',
            'H': 'Hypertension',
            'M': 'Pathological_Myopia',
            'O': 'Other_diseases'
        }
        
        self.disease_columns = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
        
        # Setup logging
        self._setup_logging()
        
        # Load and validate data
        self.df = self._load_and_validate_data()
        
        # Statistics
        self.stats = {
            'total_patients': 0,
            'total_images': 0,
            'disease_distribution': {},
            'multi_disease_patients': 0,
            'missing_images': [],
            'processed_images': 0
        }
    
    def _setup_logging(self):
        """Setup logging configuration"""
        log_dir = self.output_base_dir / 'logs'
        log_dir.mkdir(parents=True, exist_ok=True)
        
        logging.basicConfig(
            level=logging.INFO,
            format='%(asctime)s - %(levelname)s - %(message)s',
            handlers=[
                logging.FileHandler(log_dir / f'mapping_{datetime.now().strftime("%Y%m%d_%H%M%S")}.log'),
                logging.StreamHandler()
            ]
        )
        self.logger = logging.getLogger(__name__)
    
    def _load_and_validate_data(self) -> pd.DataFrame:
        """Load CSV data and perform basic validation"""
        try:
            df = pd.read_csv(self.csv_path)
            self.logger.info(f"Loaded dataset with {len(df)} records")
            
            # Validate required columns
            required_cols = ['ID', 'Left-Fundus', 'Right-Fundus'] + self.disease_columns
            missing_cols = [col for col in required_cols if col not in df.columns]
            if missing_cols:
                raise ValueError(f"Missing required columns: {missing_cols}")
            
            # Fill NaN values in disease columns with 0
            df[self.disease_columns] = df[self.disease_columns].fillna(0)
            
            return df
            
        except Exception as e:
            self.logger.error(f"Error loading data: {str(e)}")
            raise
    
    def analyze_dataset(self) -> Dict:
        """Analyze dataset and generate statistics"""
        self.logger.info("Analyzing dataset...")
        
        # Basic statistics
        self.stats['total_patients'] = len(self.df)
        self.stats['total_images'] = len(self.df) * 2  # left + right
        
        # Disease distribution
        for col in self.disease_columns:
            disease_name = self.disease_mapping[col]
            count = (self.df[col] == 1).sum()
            self.stats['disease_distribution'][disease_name] = count
        
        # Multi-disease patients
        disease_sums = self.df[self.disease_columns].sum(axis=1)
        self.stats['multi_disease_patients'] = (disease_sums > 1).sum()
        
        # Check for missing images
        for idx, row in self.df.iterrows():
            left_img = self.source_image_dir / row['Left-Fundus']
            right_img = self.source_image_dir / row['Right-Fundus']
            
            if not left_img.exists():
                self.stats['missing_images'].append(str(left_img))
            if not right_img.exists():
                self.stats['missing_images'].append(str(right_img))
        
        self.logger.info(f"Dataset Analysis Complete:")
        self.logger.info(f"  - Total Patients: {self.stats['total_patients']}")
        self.logger.info(f"  - Multi-disease Patients: {self.stats['multi_disease_patients']}")
        self.logger.info(f"  - Missing Images: {len(self.stats['missing_images'])}")
        
        return self.stats
    
    def create_single_label_mapping(self, priority_order: Optional[List[str]] = None) -> None:
        """
        Create single-label classification folders (each image goes to one folder)
        Based on priority order for patients with multiple diseases
        
        Args:
            priority_order: List of disease codes in priority order. 
                          Default: ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
        """
        if priority_order is None:
            priority_order = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
        
        self.logger.info("Creating single-label classification folders...")
        
        # Create directory structure
        single_label_dir = self.output_base_dir / 'single_label_classification'
        for disease_code in priority_order:
            disease_name = self.disease_mapping[disease_code]
            (single_label_dir / disease_name / 'left_eye').mkdir(parents=True, exist_ok=True)
            (single_label_dir / disease_name / 'right_eye').mkdir(parents=True, exist_ok=True)
        
        # Process each patient
        mapping_log = []
        
        for idx, row in self.df.iterrows():
            # Determine primary disease based on priority
            primary_disease = self._get_primary_disease(row, priority_order)
            disease_folder = self.disease_mapping[primary_disease]
            
            # Copy images
            patient_id = row['ID']
            left_img = row['Left-Fundus']
            right_img = row['Right-Fundus']
            
            # Copy left eye image
            src_left = self.source_image_dir / left_img
            dst_left = single_label_dir / disease_folder / 'left_eye' / left_img
            
            if src_left.exists():
                shutil.copy2(src_left, dst_left)
                self.stats['processed_images'] += 1
            else:
                self.logger.warning(f"Missing left image: {src_left}")
            
            # Copy right eye image
            src_right = self.source_image_dir / right_img
            dst_right = single_label_dir / disease_folder / 'right_eye' / right_img
            
            if src_right.exists():
                shutil.copy2(src_right, dst_right)
                self.stats['processed_images'] += 1
            else:
                self.logger.warning(f"Missing right image: {src_right}")
            
            # Log mapping
            diseases_present = [self.disease_mapping[col] for col in self.disease_columns if row[col] == 1]
            mapping_log.append({
                'patient_id': patient_id,
                'primary_disease': self.disease_mapping[primary_disease],
                'all_diseases': diseases_present,
                'left_image': left_img,
                'right_image': right_img
            })
        
        # Save mapping log
        self._save_mapping_log(mapping_log, single_label_dir / 'mapping_log.json')
        self.logger.info(f"Single-label mapping complete. Processed {self.stats['processed_images']} images.")
    
    def create_multi_label_mapping(self) -> None:
        """
        Create multi-label classification folders (images copied to multiple folders)
        Each image is copied to all relevant disease folders
        """
        self.logger.info("Creating multi-label classification folders...")
        
        # Create directory structure
        multi_label_dir = self.output_base_dir / 'multi_label_classification'
        for disease_code in self.disease_columns:
            disease_name = self.disease_mapping[disease_code]
            (multi_label_dir / disease_name / 'left_eye').mkdir(parents=True, exist_ok=True)
            (multi_label_dir / disease_name / 'right_eye').mkdir(parents=True, exist_ok=True)
        
        # Process each patient
        mapping_log = []
        processed_count = 0
        
        for idx, row in self.df.iterrows():
            patient_id = row['ID']
            left_img = row['Left-Fundus']
            right_img = row['Right-Fundus']
            
            # Get all diseases for this patient
            patient_diseases = [col for col in self.disease_columns if row[col] == 1]
            
            # Copy images to all relevant disease folders
            for disease_code in patient_diseases:
                disease_folder = self.disease_mapping[disease_code]
                
                # Copy left eye image
                src_left = self.source_image_dir / left_img
                dst_left = multi_label_dir / disease_folder / 'left_eye' / left_img
                
                if src_left.exists():
                    shutil.copy2(src_left, dst_left)
                    processed_count += 1
                
                # Copy right eye image
                src_right = self.source_image_dir / right_img
                dst_right = multi_label_dir / disease_folder / 'right_eye' / right_img
                
                if src_right.exists():
                    shutil.copy2(src_right, dst_right)
                    processed_count += 1
            
            # Log mapping
            diseases_present = [self.disease_mapping[col] for col in patient_diseases]
            mapping_log.append({
                'patient_id': patient_id,
                'diseases': diseases_present,
                'left_image': left_img,
                'right_image': right_img
            })
        
        # Save mapping log
        self._save_mapping_log(mapping_log, multi_label_dir / 'mapping_log.json')
        self.logger.info(f"Multi-label mapping complete. Processed {processed_count} image copies.")
    
    def create_patient_level_mapping(self) -> None:
        """
        Create patient-level folders keeping left and right eye images together
        """
        self.logger.info("Creating patient-level mapping...")
        
        patient_dir = self.output_base_dir / 'patient_level_mapping'
        
        # Create disease folders
        for disease_code in self.disease_columns:
            disease_name = self.disease_mapping[disease_code]
            (patient_dir / disease_name).mkdir(parents=True, exist_ok=True)
        
        mapping_log = []
        
        for idx, row in self.df.iterrows():
            patient_id = row['ID']
            left_img = row['Left-Fundus']
            right_img = row['Right-Fundus']
            
            # Get primary disease (for folder assignment)
            primary_disease = self._get_primary_disease(row)
            disease_folder = self.disease_mapping[primary_disease]
            
            # Create patient subfolder
            patient_folder = patient_dir / disease_folder / f"patient_{patient_id}"
            patient_folder.mkdir(exist_ok=True)
            
            # Copy both images to patient folder
            src_left = self.source_image_dir / left_img
            src_right = self.source_image_dir / right_img
            
            if src_left.exists():
                shutil.copy2(src_left, patient_folder / left_img)
            if src_right.exists():
                shutil.copy2(src_right, patient_folder / right_img)
            
            # Create patient metadata
            patient_metadata = {
                'patient_id': patient_id,
                'age': row.get('Patient Age', 'Unknown'),
                'sex': row.get('Patient Sex', 'Unknown'),
                'diseases': {disease_code: int(row[disease_code]) for disease_code in self.disease_columns},
                'left_keywords': row.get('Left-Diagnostic Keywords', ''),
                'right_keywords': row.get('Right-Diagnostic Keywords', ''),
                'left_image': left_img,
                'right_image': right_img
            }
            
            # Save patient metadata
            with open(patient_folder / 'metadata.json', 'w') as f:
                json.dump(patient_metadata, f, indent=2)
            
            mapping_log.append(patient_metadata)
        
        # Save overall mapping log
        self._save_mapping_log(mapping_log, patient_dir / 'patient_mapping_log.json')
        self.logger.info("Patient-level mapping complete.")
    
    def create_train_val_test_split(self, train_ratio: float = 0.7, val_ratio: float = 0.15, 
                                   test_ratio: float = 0.15, random_state: int = 42) -> None:
        """
        Create train/validation/test splits at patient level to avoid data leakage
        """
        self.logger.info("Creating train/validation/test splits...")
        
        if abs(train_ratio + val_ratio + test_ratio - 1.0) > 1e-6:
            raise ValueError("Split ratios must sum to 1.0")
        
        # Get unique patient IDs
        patient_ids = self.df['ID'].unique()
        np.random.seed(random_state)
        np.random.shuffle(patient_ids)
        
        # Calculate split indices
        n_patients = len(patient_ids)
        train_end = int(n_patients * train_ratio)
        val_end = train_end + int(n_patients * val_ratio)
        
        # Split patient IDs
        train_patients = patient_ids[:train_end]
        val_patients = patient_ids[train_end:val_end]
        test_patients = patient_ids[val_end:]
        
        # Create split directories
        split_dir = self.output_base_dir / 'train_val_test_split'
        for split in ['train', 'validation', 'test']:
            for disease_code in self.disease_columns:
                disease_name = self.disease_mapping[disease_code]
                (split_dir / split / disease_name / 'left_eye').mkdir(parents=True, exist_ok=True)
                (split_dir / split / disease_name / 'right_eye').mkdir(parents=True, exist_ok=True)
        
        # Process splits
        split_mapping = {'train': train_patients, 'validation': val_patients, 'test': test_patients}
        split_stats = {}
        
        for split_name, patient_list in split_mapping.items():
            split_df = self.df[self.df['ID'].isin(patient_list)]
            split_stats[split_name] = {'patients': len(patient_list), 'images': len(split_df) * 2}
            
            for idx, row in split_df.iterrows():
                # Get primary disease
                primary_disease = self._get_primary_disease(row)
                disease_folder = self.disease_mapping[primary_disease]
                
                # Copy images
                left_img = row['Left-Fundus']
                right_img = row['Right-Fundus']
                
                src_left = self.source_image_dir / left_img
                src_right = self.source_image_dir / right_img
                
                dst_left = split_dir / split_name / disease_folder / 'left_eye' / left_img
                dst_right = split_dir / split_name / disease_folder / 'right_eye' / right_img
                
                if src_left.exists():
                    shutil.copy2(src_left, dst_left)
                if src_right.exists():
                    shutil.copy2(src_right, dst_right)
        
        # Save split information
        split_info = {
            'split_ratios': {'train': train_ratio, 'validation': val_ratio, 'test': test_ratio},
            'split_stats': split_stats,
            'random_state': random_state
        }
        
        with open(split_dir / 'split_info.json', 'w') as f:
            json.dump(split_info, f, indent=2)
        
        self.logger.info(f"Train/val/test split complete:")
        for split_name, stats in split_stats.items():
            self.logger.info(f"  {split_name}: {stats['patients']} patients, {stats['images']} images")
    
    def _get_primary_disease(self, row: pd.Series, priority_order: Optional[List[str]] = None) -> str:
        """Get primary disease based on priority order"""
        if priority_order is None:
            priority_order = ['N', 'D', 'G', 'C', 'A', 'H', 'M', 'O']
        
        for disease_code in priority_order:
            if row[disease_code] == 1:
                return disease_code
        
        # Fallback to 'Other' if no disease is marked
        return 'O'
    
    def _save_mapping_log(self, mapping_log: List[Dict], filepath: Path) -> None:
        """Save mapping log to JSON file"""
        with open(filepath, 'w') as f:
            json.dump(mapping_log, f, indent=2)
        self.logger.info(f"Mapping log saved to {filepath}")
    
    def generate_summary_report(self) -> None:
        """Generate a comprehensive summary report"""
        report_path = self.output_base_dir / 'mapping_summary_report.txt'
        
        with open(report_path, 'w') as f:
            f.write("EYE DISEASE DATASET MAPPING SUMMARY REPORT\n")
            f.write("=" * 50 + "\n\n")
            
            f.write(f"Dataset Overview:\n")
            f.write(f"  Total Patients: {self.stats['total_patients']}\n")
            f.write(f"  Total Images: {self.stats['total_images']}\n")
            f.write(f"  Multi-disease Patients: {self.stats['multi_disease_patients']}\n")
            f.write(f"  Missing Images: {len(self.stats['missing_images'])}\n\n")
            
            f.write("Disease Distribution:\n")
            for disease, count in self.stats['disease_distribution'].items():
                percentage = (count / self.stats['total_patients']) * 100
                f.write(f"  {disease}: {count} ({percentage:.1f}%)\n")
            
            f.write(f"\nProcessed Images: {self.stats['processed_images']}\n")
            f.write(f"Processing Date: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}\n")
        
        self.logger.info(f"Summary report saved to {report_path}")


def main():
    """Example usage of the EyeDiseaseMapper"""
    
    # Configuration
    CSV_PATH = "/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/full_df.csv"
    SOURCE_IMAGE_DIR = "/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/raw_data/combine_all_image"  # Directory containing all original images
    OUTPUT_BASE_DIR = "/Volumes/KODAK/folder_02/eye_disease_prediction_model/data/preprocess_data"
    
    # Initialize mapper
    mapper = EyeDiseaseMapper(CSV_PATH, SOURCE_IMAGE_DIR, OUTPUT_BASE_DIR)
    
    # Analyze dataset
    stats = mapper.analyze_dataset()
    
    # Create different mapping structures
    print("Creating single-label classification folders...")
    mapper.create_single_label_mapping()
    
    print("Creating multi-label classification folders...")
    mapper.create_multi_label_mapping()
    
    print("Creating patient-level mapping...")
    mapper.create_patient_level_mapping()
    
    print("Creating train/validation/test splits...")
    mapper.create_train_val_test_split()
    
    # Generate summary report
    mapper.generate_summary_report()
    
    print(f"Mapping complete! Check the '{OUTPUT_BASE_DIR}' directory for organized data.")


if __name__ == "__main__":
    main()
```
